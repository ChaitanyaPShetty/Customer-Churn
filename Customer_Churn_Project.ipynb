## FEATURE SELECTION

```python
def add_change_kpis_from_6_7_to_8(df, feature_names):
    for f_name in feature_names:
        df[f_name+'_6'] = df[f_name+'_6'].fillna(0) # Imputing missing values with 0
        df[f_name+'_7'] = df[f_name+'_7'].fillna(0) # Imputing missing values with 0
        df[f_name+'_8'] = df[f_name+'_8'].fillna(0)# Imputing missing values with 0
        df['change_in_'+f_name] = (df[f_name+'_6'] + df[f_name+'_7'])/2 - df[f_name+'_8']
    return df
```

```python
df_processed = add_change_kpis_from_6_7_to_8(dataset,['arpu','onnet_mou','offnet_mou','total_og_mou','total_ic_mou','total_rech_amt',
                                                  'spl_ic_mou','spl_og_mou','total_rech_num','total_rech_data','count_rech_2g',
                                                            'count_rech_3g','arpu_3g','arpu_2g','monthly_2g','sachet_2g','monthly_3g',
                                                            'sachet_3g','av_rech_amt_data', 'roam_ic_mou','roam_og_mou','loc_og_t2t_mou',
                                                           'loc_og_t2m_mou','loc_og_t2f_mou','loc_og_t2c_mou','loc_og_mou','std_og_t2t_mou',
                                                           'std_og_t2m_mou','std_og_t2f_mou','std_og_mou','isd_og_mou','og_others',
                                                           'loc_ic_t2t_mou','loc_ic_t2m_mou','loc_ic_t2f_mou','loc_ic_mou','std_ic_t2t_mou',
                                                           'std_ic_t2m_mou','std_ic_t2f_mou','std_ic_t2o_mou','std_ic_mou','isd_ic_mou',
                                                           'ic_others','night_pck_user','max_rech_amt','last_day_rch_amt','max_rech_data',
                                                           'av_rech_amt_data'])

#Handling special scenario where column names are not in standard format
df_processed['jul_vbc_3g'] = df_processed['jul_vbc_3g'].fillna(0)
df_processed['aug_vbc_3g'] = df_processed['aug_vbc_3g'].fillna(0)
df_processed['jun_vbc_3g'] = df_processed['jun_vbc_3g'].fillna(0)
df_processed['change_in_vbc_3g'] = (df_processed['jul_vbc_3g'] + df_processed['jun_vbc_3g']) /2 - df_processed['aug_vbc_3g']
```

```python
dataset['total_data_rech_6']= dataset['total_rech_data_6'] * dataset['av_rech_amt_data_6']
dataset['total_data_rech_7']= dataset['total_rech_data_7'] * dataset['av_rech_amt_data_7']
dataset['total_data_rech_8']= dataset['total_rech_data_8'] * dataset['av_rech_amt_data_8']
dataset['total_data_rech_6'] = dataset['total_data_rech_6'].fillna(0) # Imputing missing values with 0
dataset['total_data_rech_7'] = dataset['total_data_rech_7'].fillna(0) # Imputing missing values with 0
dataset['total_data_rech_8'] = dataset['total_data_rech_8'].fillna(0) # Imputing missing values with 0

dataset['amt_data_6'] = dataset['total_rech_amt_6'] + dataset['total_data_rech_6']
dataset['amt_data_7'] = dataset['total_rech_amt_7'] + dataset['total_data_rech_7']
dataset['amt_data_8'] = dataset['total_rech_amt_8'] + dataset['total_data_rech_8']
```

```python
def get_difference_in_days(date_co1, date_col2):
    return (pd.to_datetime(date_co1)- pd.to_datetime(date_col2))/np.timedelta64(1,'D')

#Derived Features after operating on dates - average duration change in last recharge of month
june_last_recharge_duration = get_difference_in_days(df_processed['last_date_of_month_6'], df_processed['date_of_last_rech_6'])
july_last_recharge_duration = get_difference_in_days(df_processed['last_date_of_month_7'], df_processed['date_of_last_rech_7'])
aug_last_recharge_duration = get_difference_in_days(df_processed['last_date_of_month_8'], df_processed['date_of_last_rech_8'])
df_processed['change_in_duration_of_last_recharge'] = (june_last_recharge_duration + july_last_recharge_duration) /2 - aug_last_recharge_duration
df_processed['change_in_duration_of_last_recharge'] = df_processed['change_in_duration_of_last_recharge'].fillna(0)

#Derived Features after operating on dates - average duration change in last data recharge of month
june_last_data_recharge_duration = get_difference_in_days(df_processed['last_date_of_month_6'], df_processed['date_of_last_rech_data_6'])
july_last_data_recharge_duration = get_difference_in_days(df_processed['last_date_of_month_6'], df_processed['date_of_last_rech_data_7'])
aug_last_data_recharge_duration = get_difference_in_days(df_processed['last_date_of_month_6'], df_processed['date_of_last_rech_data_8'])
df_processed['change_in_duration_of_last_data_recharge'] = (june_last_data_recharge_duration + july_last_data_recharge_duration) /2 - aug_last_data_recharge_duration
df_processed['change_in_duration_of_last_data_recharge'] = df_processed['change_in_duration_of_last_data_recharge'].fillna(0)
```
